1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np


x = sp.symbols('x')

f_lin = 10 * x -5
derivative_f_lin = sp.diff(f_lin, x)

print("Похідна лінійної функції 10x-5:", derivative_f_lin)
```

```python
def f(x):
  return 10*x-5

xs = np.linspace(0,10,20)
plt.plot(xs, f(xs), 'o');
```

```python
# Створення масиву x
x = np.linspace(0, 10, 100)
f = 10 * x -5
df = np.full_like(x, 10)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10 * x -5)', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Чим більше ти навчаєшся тим вищу оцінку отримаєш. Похідна функції f'(x) = 10 є сталою, що означає, що кожна додаткова година навчання збільшує оцінку на 10 балів.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x = sp.symbols('x')
f_sin=sp.sin(x)
derivative_f_sin=sp.diff(f_sin, x)
print("Похідна  функції sin(x):", derivative_f_sin)
```

```python
x = sp.symbols('x')
f_ln=sp.ln(x)
derivative_f_ln=sp.diff(f_ln, x)
print("Похідна  функції ln(x):", derivative_f_ln)
```

```python
# Створення масиву x
x = np.linspace(0, 10, 100)
f = np.sin(x)
df = np.cos(x)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = sin(x)', color='blue')
plt.plot(x, df, label="f'(x) = cos(x)", color='green')

# Позначення точки та дотичної лінії
x_point = 3
f_point = np.sin(x_point)
df_point = np.cos(x_point)
tangent_line = df_point * (x - x_point) + f_point  # Рівняння дотичної: y - y1 = m(x - x1)

plt.scatter(x_point, f_point, color='red') # точка на кривій
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 3') # дотична

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Створення масиву x
x = np.linspace(0.1, 5, 100)
f = np.log(x)
df = 1/x

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = ln(x)', color='blue')
plt.plot(x, df, label="f'(x) = 1/x", color='green')

# Позначення точки та дотичної лінії
x_point = 4
f_point = np.log(x_point)
df_point = 1/x_point
tangent_line = df_point * (x - x_point) + f_point  # Рівняння дотичної: y - y1 = m(x - x1)

plt.scatter(x_point, f_point, color='red') # точка на кривій
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 4') # дотична

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
f = x**2-5*x+1
derivative_f = sp.diff(f, x)
print("Похідна  функції x2−5x+1:", derivative_f)
```

```python
# Створення масиву x
x = np.linspace(0, 100, 100)
f = x**2-5*x+1
df = 2*x - 5

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = x^2−5x+1', color='blue')
plt.plot(x, df, label="f'(x) = 2x - 5", color='green')

# Позначення точки та дотичної лінії
x_point = 60
f_point = (x_point)**2 - 5*(x_point)+1
df_point = 2*(x_point)-5
tangent_line = df_point * (x - x_point) + f_point  # Рівняння дотичної: y - y1 = m(x - x1)

plt.scatter(x_point, f_point, color='red') # точка на кривій
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 60') # дотична

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
